# Part 1 - Data Preprocessing

Importing the libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

Importing the dataset

```python
#獲取當前路徑
import os
os.chdir('/Users/shiliu/Downloads')
os.getcwd()

dataset = pd.read_csv('Churn_Modelling.csv')
dataset
```

```python
X = dataset.iloc[:, 3:13].values
# take all the rows
# take all the CreditScore to EstimatedSalary
y = dataset.iloc[:, 13].values
# take all the rows
# vector
pd.DataFrame(X)
```

Encoding categorical data
- categorical data, ex: country

```python
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
# First categorical variable : country 
labelencoder_X_1 = LabelEncoder()
X[:, 1] = labelencoder_X_1.fit_transform(X[:, 1])
     #1 means the first column
    
# Second categorical variable : country 
labelencoder_X_2 = LabelEncoder()
X[:, 2] = labelencoder_X_2.fit_transform(X[:, 2])
     #2 means the second column

onehotencoder = OneHotEncoder(categorical_features=[1])
X = onehotencoder.fit_transform(X).toarray()
# Update the data to avoid the dummy variable trap
# taking all lines(rows) and all the columns except the first one
# index 1 corresponds to the second column
X = X[:, 1:]
pd.DataFrame(X)
```

Split the dataset into a training set and a test set

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, 
                                                     test_size = 0.2, 
                                                     random_state = 0)
#test_size = 0.2 = 20%
```

Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
pd.DataFrame(X_train)
```

# Part 2 - Now let's make the ANN !

```python
import keras
from keras.models import Sequential
from keras.layers import Dense

# Initialising the ANN
classifier = Sequential()
# Adding the input layer and the first hidden layer
print(X_train.shape)
       #input layers are 11 and the output layer is 1, so the mean is 6
classifier.add(Dense(output_dim = 6, init = 'uniform', 
                     activation = 'relu', input_dim = 11))
# Adding the second hidden layer
classifier.add(Dense(output_dim = 6, init = 'uniform', 
                     activation = 'relu'))
# Adding the output layer
classifier.add(Dense(output_dim = 1, init = 'uniform',
                     activation = 'sigmoid'))
# Compiling the ANN
classifier.compile(optimizer = 'adam', loss = 'binary_crossentropy',
                  metrics = ['accuracy'])
# Fitting the ANN to the Training set
classifier.fit(X_train, y_train, batch_size = 10, nb_epoch = 100)
```

# Part 3 - Making the predictions and evaluating the model

Predicting the Test set results

```python
y_pred = classifier.predict(X_test)
y_pred = (y_pred > 0.5)
pd.DataFrame(y_pred)
 # false : dont leave the bank 
```

Making the Confusion Matrix

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
pd.DataFrame(cm)
```
